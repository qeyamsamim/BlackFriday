```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
data = pd.read_csv('input/BlackFriday.csv')
```

```python
data
```

```python
data1 = data.copy()
```

```python
data1 = data1.groupby('Product_ID')['Purchase'].agg(['sum', 'count'])
data1.rename(columns={'sum': 'Total Amount', 'count': 'Total Purchase'}, inplace=True)
data1
```

```python
data1.nlargest(5, 'Total Purchase')
```

```python
data1.loc[data1['Total Amount'].idxmax()]
```

```python
lst=[]
for item in data['User_ID'].unique():
    lst2=list(set(data[data['User_ID']==item]['Product_ID']))
    if len(lst2)>0:
        lst.append(lst2)

items = TransactionEncoder()
items_ary = items.fit(lst).transform(lst)
data = pd.DataFrame(items_ary, columns=items.columns_)
data
```

```python
frequent_itemsets = apriori(data, min_support=0.02, use_colnames=True)
```

```python
frequent_itemsets
```

```python
frequent_itemsets.nlargest(5, 'support')
```

```python
from mlxtend.frequent_patterns import association_rules
data = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
```
